```python
import math
import numpy as np

A = [[7,32,-83,47,-52],
     [95,-41,-13,-197,65],
     [-34,-64,31,36,-14],
     [4,-66,-57,-66,-100],
     [112,-23,54,-28,18]]

AT = np.transpose(A)
```

```python
def gradient_descent(oldSol, gamma, grad):

    newSol = oldSol- gamma * grad

    return np.array(newSol)

def CalculateGrad(U):

    Grad = 2 * np.matmul(np.matmul(AT,A),U) + 2* np.matmul(U,np.matmul(A,AT)) - 4* AT

    return Grad

def CalculateLoss(U):

    Loss = np.linalg.norm(np.matmul(A,U)-np.identity(5), ord='fro') + np.linalg.norm(np.matmul(U,A)-np.identity(5), ord='fro')

    return Loss
```

```python
def GD(gamma):

    LVals = []

    U0 = np.identity(5)
    # U0 = A
    LVals.append(CalculateLoss(U0))

    grad = CalculateGrad(U0)

    oldSol = gradient_descent(U0, gamma, grad)

    LVals.append(CalculateLoss(oldSol))

    iterCount = 1

    while iterCount <= 100:

        # print(iterCount)

        grad = CalculateGrad(oldSol)
        oldSol = gradient_descent(oldSol, gamma, grad)
        LVals.append(CalculateLoss(oldSol))

        iterCount +=1

    # print(LVals)

    print(oldSol)

    return LVals
```

### Gradient Decent Function

```python

# Gradient descent function with loss tracking
def gradient_descent_c(m,u,initial_x, learning_rate, num_iterations):
    x = initial_x
    losses = []
    results = []
    for i in range(num_iterations):
        loss = func(x)  # Calculate loss at current x
        losses.append(loss)
        grad = expect(m,u,x)  # Calculate gradient of f at current x
        x = x - learning_rate * grad  # Update x using gradient descent update rule
        results.append(x)
    return losses
```

```python
import numpy as np
import math
import csv


def lintrans(CoefMatrix,Coefvector,x):
    Wx = np.matmul(CoefMatrix,x)
    return [sum(t) for t in zip(Wx, Coefvector)]

def ReLu(z):
    zprime = []

    for z_i in z:
        if z_i >= 0:
            zprime.append(z_i)

        else:
            zprime.append(np.array([0]))

    return zprime


def softmax(y):
    y1prime = math.exp(y[0]) / (math.exp(y[0]) + math.exp(y[1]))
    y2prime = math.exp(y[1]) / (math.exp(y[0]) + math.exp(y[1]))
    return np.transpose([y1prime, y2prime])

def CEL(yprime,ytilda): ## Cross Entropy Loss
    return -ytilda*math.log(yprime[0]) - (1-ytilda)*math.log(yprime[1])

def Loss(CEs,numofSamples):
    return 1/numofSamples* sum(CEs)


def gradLW(delta2,x):

    delta2 = np.reshape(delta2,(1,len(delta2)))
    x = np.reshape(x,(len(x),1))

    return(np.matmul(np.transpose(delta2),np.transpose(x)))

def gradLWprime(delta1,h):
    delta1= np.reshape(delta1,(1,len(delta1)))
    h = np.reshape(h,(len(h),1))

    return(np.matmul(np.transpose(delta1),np.transpose(h)))

def gradLbprime(delta1):
    delta1= np.reshape(delta1,(1,len(delta1)))
    return np.transpose(delta1)

def gradLb(delta2):
    delta2 = np.reshape(delta2,(1,len(delta2)))
    return np.transpose(delta2)

def find_delta1(yprime,ytilda):
    return [yprime[0]-ytilda, yprime[1]+ytilda-1]

def find_delta2(delta1,Wprime,z):
    delta2 = np.matmul(delta1,Wprime)

    for i in range (0,len(delta2)):
        if z[i] < 0:
            delta2[i] = 0

    return delta2


def NN(x,ytilda,W,Wprime,b,bprime):
    n = len(ytilda)
    CEs = []

    dLW = np.zeros((1000,2))
    dLWprime = np.zeros((2,1000))
    dLb = np.zeros((1000,1))
    dLbprime = np.zeros((2,1))

    for sample in range(n):
        z = lintrans(W,b,x[sample])
        zprime = ReLu(z)
        y = lintrans(Wprime,bprime,zprime)
        yprime = softmax(y)
        CE_sample = CEL(yprime,ytilda[sample])
        CEs.append(CE_sample)

        delta1 = find_delta1(yprime,ytilda[sample])
        delta2 = find_delta2(delta1,Wprime,z)

        dLW += gradLW(delta2,x[sample])
        dLWprime += gradLWprime(delta1,zprime)
        dLb += gradLb(delta2)
        dLbprime += gradLbprime(delta1)

    LossValue = Loss(CEs,n)

    return LossValue, dLW, dLWprime, dLb, dLbprime

def gradientdescent(inputs,ytildas):
    W = np.random.randn(1000,2)* 1/100
    Wprime = np.random.randn(2,1000)* 1/100
    b = np.zeros((1000,1))
    bprime = np.zeros((2,1))

    # W = np.zeros((1000,2))* 1/25
    # Wprime = np.zeros((2,1000))* 1/25
    # b = np.zeros((1000,1))
    # bprime = np.zeros((2,1))

    Losses = []

    gamma = 1.041610

    itera = 0

    while itera <= 4000:

        Loss, dLW, dLWprime, dLb, dLbprime = NN(inputs,ytildas,W,Wprime,b,bprime)

        Losses.append(Loss)

        if itera <= 3999:

            W -= gamma * dLW/len(ytildas)
            Wprime -= gamma * dLWprime/len(ytildas)
            b -= gamma * dLb/len(ytildas)
            bprime -= gamma * dLbprime/len(ytildas)

        itera +=1

    return Losses, W,Wprime,b,bprime
```
